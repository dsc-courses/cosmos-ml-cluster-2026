```python
import re

def tokenize(review):
  review = review.lower()
  split = re.findall(r'\w+', review)
  return split[0]

# testing BagOfWordsByClass[c]
def wordOccurence(word, X, y):
  count = 0
  countpos = 0
  countneg = 0
  for i in range(len(X)):
    if tokenize(word) in X[i]:
      count += 1
      if y[i] == 0:
        countneg += 1
      else:
        countpos += 1
  #print(count, "occurrences of", word, "in all reviews")
  print(countpos, "occurrences of", word, "in positive reviews")
  print(countneg, "occurrences of", word, "in negative reviews")

def testBagOfWordsByClass(word, X, y, bagOfWords):
  #print("* is word the same before and after tokenization?") # is the word same before and after tokenization?
  #if (tokenize(word)[0] == word):
  #  print("Yes")
  #else:
  #  print("No")
  #  print("tokenized word:", tokenize(word))

  print("--- in original dataset ---")
  wordOccurence(word, X, y)
  print("--- in BagOfWordsByClass ---")
  print(bagOfWords[1].get(word), "occurrences of", word, "in positive reviews")
  print(bagOfWords[0].get(word), "occurrences of", word, "in negative reviews")
```
